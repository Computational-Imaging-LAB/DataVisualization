```

import numpy as np

 
plt.style.use('seaborn-darkgrid')
palette = plt.get_cmap('Set1')
plt.figure(figsize=(16,16))


def lineplotter(df,suptitle,plotDict,focus=None):
  
  for num,column in enumerate(df):
      
      cmap=plotDict['palette']
      if focus ==None:
        ax=plt.subplot(3,3, num+1)

        ax.plot(df[column], marker=plotDict['marker'], color=cmap(num), linewidth=plotDict['linewidth'], alpha=plotDict['alpha'], label=column)
        plt.title(column, loc='left', fontsize=suptitle['fontsize'], fontweight=suptitle['fontweight'], color=cmap(num) )
      else:
        ax=plt.subplot(1,1,1)

        if column in focus['name']:
          ax.plot(df[column], marker=plotDict['marker'], color=cmap(num), linewidth=plotDict['linewidth'], alpha=plotDict['alpha'], label=column)
          plt.title(column, loc='left', fontsize=suptitle['fontsize'], fontweight=suptitle['fontweight'], color=cmap(num) )
          plt.text((df[column].index[-1]), df[column].tail(1), column, horizontalalignment='left', size='large', color=cmap(num))
          
        else:
          ax.plot(df[column], marker=plotDict['marker'], color='gray', linewidth=plotDict['linewidth'], alpha=0.2, label=column)
          plt.title(column, loc='left', fontsize=suptitle['fontsize'], fontweight=suptitle['fontweight'], color=cmap(num) )
          loc,_=plt.xticks()
          
          plt.text((df[column].index[-1]), df[column].tail(1), column, horizontalalignment='left', size='small', color='grey')
 


  plt.suptitle("Subplots of features", fontsize=13, fontweight=0, color='black', style='italic', y=1.12)
  


df2=df.drop('origin', axis=1)
df2=df.iloc[df.groupby('origin').groups['europe'],:] # we choose only europe

suptitle={'fontsize':13,'fontweight':0,'color':'black','style':'italic'}
plotDict={'marker':'','palette':plt.get_cmap('Set1'),'linewidth':2,'alpha':1}

lineplotter(df2,suptitle=suptitle,plotDict=plotDict)
```

```
df2=df.iloc[df.groupby('origin').groups['europe'],:] # we choose only europe
df2=df2[['horsepower','acceleration','mpg','displacement']]
fig=plt.figure(figsize=(8,8))

suptitle={'fontsize':13,'fontweight':0,'color':'black','style':'italic'}
plotDict={'marker':'','palette':plt.get_cmap('Set1'),'linewidth':2,'alpha':1}

lineplotter(df2,suptitle=suptitle,plotDict=plotDict,focus={'name':['mpg','acceleration']})
plt.legend()
```

```
g = sns.FacetGrid(dfA, col='origin', hue='origin', col_wrap=3, )
 
# Add the line over the area with the plot function
g = g.map(plt.plot, 'model_year', 'weight')
 
# Fill the area with fill_between
g = g.map(plt.fill_between, 'model_year', 'weight', alpha=0.2).set_titles("{col_name} model_year")
 
# Control the title of each facet
g = g.set_titles("{col_name}")
 
# Add a title for the whole plo
plt.subplots_adjust(top=0.92)
g = g.fig.suptitle('Car Weights among countries ',y=1.25)
 
plt.show()


g = sns.FacetGrid(dfA, col='origin', hue='origin', col_wrap=3, )
 
# Add the line over the area with the plot function
g = g.map(plt.plot, 'model_year', 'acceleration')
 
# Fill the area with fill_between
g = g.map(plt.fill_between, 'model_year', 'acceleration', alpha=0.2).set_titles("{col_name} model_year")
 
# Control the title of each facet
g = g.set_titles("{col_name}")
 
# Add a title for the whole plo
plt.subplots_adjust(top=0.92)
g = g.fig.suptitle('Acceleration among countries ',y=1.25)
 
plt.show()
```
