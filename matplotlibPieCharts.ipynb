```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df=sns.load_dataset('iris')

grouper=df.groupby('species')

sz=(grouper.sepal_length.count())


plt.pie(sz)
#plt.show()
 
# add a circle at the center
my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
 
plt.show()
```

```
plt.pie(sz, labels=pd.unique(df.species), wedgeprops = { 'linewidth' : 7, 'edgecolor' : 'white' }, colors=['red','green','skyblue'])

plt.show()
plt.pie(sz, labels=pd.unique(df.species), wedgeprops = { 'linewidth' : 7, 'edgecolor' : 'white' })
p=plt.gcf()
my_circle=plt.Circle( (0,0), 0.7, color='white')
p.gca().add_artist(my_circle)
plt.show()
```

```
pd.Series(sz).plot.pie ( subplots=True, figsize=(16,8),autopct="%.2f",)
plt.ylabel('')
```
