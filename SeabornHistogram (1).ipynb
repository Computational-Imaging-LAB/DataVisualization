```
import seaborn as sns
df = sns.load_dataset('iris')
df
```

```
# Import library and dataset

sns.histplot( df["sepal_length"],color=[0,0,0],fill=True,bins=20,kde=True,line_kws={'color':"r", "alpha":1, "linewidth": 5} )

 
# Control the number of bins
sns.displot( df["sepal_length"])
```

```
import seaborn as sns
import matplotlib.pyplot as plt
 
f, (ax_box, ax_hist) = plt.subplots(2,gridspec_kw={'height_ratios':[0.15,0.85]})
 
sns.boxplot(x=df["sepal_length"], ax=ax_box)
sns.histplot(x=df["sepal_length"],bins=20, ax=ax_hist)
 
```
