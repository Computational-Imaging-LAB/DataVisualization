```
import seaborn as sns
df = sns.load_dataset('iris')
df
```

```
sns.jointplot(x=df["petal_length"], y=df["petal_width"], kind='scatter',color=[0,1,1],height=10)
sns.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind='hex',height=11,space=4,marginal_kws={'color':[0,0,0]})
sns.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind='hist')
sns.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind='reg')
sns.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind='resid')
sns.jointplot(x=df["sepal_length"], y=df["sepal_width"],kind='kde',shade=True)


```

```

from matplotlib import pyplot as plt

f,ax=plt.subplots(1,1)
sns.kdeplot(data=df, x=df['sepal_length'],y=df['sepal_width'],ax=ax)
f,ax=plt.subplots(1,1)

sns.kdeplot(x=df.sepal_width, y=df.sepal_length, cmap="Reds", shade=True,ax=ax)
f,ax=plt.subplots(1,1)

sns.kdeplot(y=df.sepal_width, x=df.sepal_length, cmap="Blues", shade=True, shade_lowest=True,ax=ax )
```
