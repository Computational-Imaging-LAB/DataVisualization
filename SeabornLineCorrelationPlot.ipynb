```
import seaborn as sns
df = sns.load_dataset('iris')
df
```

```
from matplotlib import pyplot as plt

f,ax=plt.subplots(1,1)

sns.regplot(ax=ax,x=df["sepal_length"], y=df["sepal_width"], line_kws={"color":"r","alpha":0.7,"lw":5},x_ci=95)
```

```

f,ax=plt.subplots(1,1)

sns.regplot(ax=ax,x=df["sepal_length"], y=df["sepal_width"], line_kws={"color":"r","alpha":0.7,"lw":5},x_ci=95,scatter_kws={"color":'skyblue',"alpha":0.5})
```

```

f,ax=plt.subplots(1,1)

sns.regplot(ax=ax,x=df["sepal_length"], y=df["sepal_width"], line_kws={"color":"r","alpha":1,"lw":1},x_ci=95,
            scatter_kws={"color":'black',"alpha":1},robust=True,ci=96,marker='*',dropna=True,x_bins=7)
```

```
sns.lmplot(x="sepal_length", y="sepal_width", hue="species", data=df)
sns.lmplot(x="sepal_length", y="sepal_width", hue="species",col="species", data=df)

sns.lmplot( x="sepal_length", y="sepal_width", data=df, fit_reg=False, hue='species', legend=False, markers=["o", "x", "1"])
 
plt.legend(loc='lower right')
sns.lmplot( x="sepal_length", y="sepal_width", data=df, fit_reg=False, hue='species', legend=False, markers=["o", "x", "1"],x_bins=1)
```

```
sns.pairplot(df, kind="reg")
 
sns.pairplot(df, palette='Oranges',kind='scatter')
```

```
sns.pairplot(df,hue='species', palette='Oranges',kind='scatter')
sns.pairplot(df, palette='PuBu',kind='reg')
```
