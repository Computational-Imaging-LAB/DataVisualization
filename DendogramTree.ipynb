```
import pandas as pd
from matplotlib import pyplot as plt
from scipy.cluster import hierarchy
import numpy as np
import seaborn as sns 
plt.figure(figsize=(16,16))
df=sns.load_dataset('iris')
 
# Calculate the distance between each sample
Z = hierarchy.linkage(df.iloc[:,:-1], 'ward')
 
 
# Plot with Custom leaves
hierarchy.dendrogram(Z, leaf_rotation=90, leaf_font_size=8, labels=df.index)
plt.xticks(rotation=90)
```

```
plt.figure(figsize=(12,12))
hierarchy.dendrogram(Z, leaf_rotation=90, orientation="right",leaf_font_size=8, labels=df.index)
```

```
df2=df 
df2= df2.set_index('species')
plt.figure(figsize=(12,12))
# Calculate the distance between each sample
Z = hierarchy.linkage(df2.iloc[:,:-1])
 
 
# Plot with Custom leaves
hierarchy.dendrogram(Z, leaf_rotation=90, leaf_font_size=10, labels=df2.index,show_leaf_counts=True,get_leaves=True,)
```
