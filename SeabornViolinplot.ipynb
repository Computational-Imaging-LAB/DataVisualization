```
# Example iris dataset load
import pandas as pd
df = sns.load_dataset('iris')
df
```

```
import seaborn as sns

#Violin plot for only one group
a=sns.violinplot( y=df["sepal_length"] )
a.set_xlabel('Seaborn Example Xlabel',fontdict=None,labelpad=4)
a.set_ylabel('Seaborn Example YLabel')
a.set_title('Seaborn Example Title',fontdict=None)
```

```
a=sns.violinplot( y=df["sepal_length"] )
a.set_facecolor([1,0,1])
a.set_alpha(2)
a.set_autoscale_on
a.set_autoscaley_on
a.set_clip_on
a.set_xbound(lower=-2,upper=2)
```

```
a=sns.violinplot( x=df['species'], y=df['sepal_length'],width=0.5 ,linewidth=4)
```

```
from matplotlib import pyplot as plt

grid = plt.GridSpec(2, 3, wspace=0.4, hspace=0.3)
ax1=plt.subplot(grid[0, 0])
sns.violinplot(ax=ax1,y=df['sepal_length'])

ax2=plt.subplot(grid[0, 1:])
a=sns.violinplot(ax=ax2,x=df['species'],y=df['sepal_length'])
a.set_xticklabels([])

ax3=plt.subplot(grid[1, :2])
a=sns.violinplot(ax=ax3,y=df['species'],x=df['sepal_length'])


ax4=plt.subplot(grid[1, 2]);
a=sns.violinplot(ax=ax4, y=df['petal_length'],x=df['species'])
a.set_xticklabels(pd.unique(df['species']),rotation=30)
```

```
#Dont worry this is actually artificial group id. I created this one because I didnt want to use another dataset.
grouper=([1]*(len(df)-130))+([2]*25)+([1]*25)+([2]*25)+([1]*25)+([2]*30)

df['grouper']=grouper
sns.violinplot(y="sepal_length",x='species', hue="grouper", data=df, palette="Pastel1")
```
