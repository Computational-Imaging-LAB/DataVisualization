```
import seaborn as sns
df = sns.load_dataset('iris')
df
```

```
#fliersize is the outlier size


ax = sns.boxplot(x='species', y='sepal_length', data=df,fliersize=15,)
ax = sns.swarmplot(x='species', y='sepal_length', data=df, color="red")
```

```
from matplotlib import pyplot as plt

f,ax=plt.subplots(1,2,gridspec_kw={'width_ratios':[0.2,0.8]})

sns.boxplot(data=df,x='species',y=df['sepal_length'],hue='species',width=0.5,notch=True,linewidth=3)

sns.boxplot(data=df,y=df['sepal_length'],hue='species',width=0.5,notch=True,linewidth=3,ax=ax[0])
```

```


ax=sns.boxplot(x=df['species'],y=df['sepal_length'])

for idx,dat in enumerate(df.groupby('species')['sepal_length'].median()):
  ax.text(idx, dat + 0.03,str(dat) ,horizontalalignment='center', size='x-small', color='w', weight='semibold')
 
```
